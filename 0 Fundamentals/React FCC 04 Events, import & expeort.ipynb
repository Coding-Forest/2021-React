**`const` and `let`**

const is a signal that the variable won't be reassigned. let is a signal that the variable may be reassigned. Additional things to ponder: Use const by default. Use let only if rebinding is needed ([StackOverflow](https://stackoverflow.com/questions/40825010/const-or-let-in-react-component#:~:text=const%20is%20a%20signal%20that,only%20if%20rebinding%20is%20needed)).
