## 3) Props (Constructors)
Let's build object constructors (create components with variable attributes.)

Compare the following: 

```python
# Static invariable constructor with hard-coded attribute values.
Javascript(""" 

function Booklist() {
  return <section className='booklist'>
      <Book/>
      <Book/>
    </section>
  }

const Book = () => {
  return (<article className='book'>
      <Image />
      <Title />     // inserted here
      <Author />    // inserted here
    </article>)
}

const Title = () => <h3>I Love you to the Moon and Back</h3>;  // hard-coded
const Author = () => <h4>Amelia Hepworth</h4>;                 // hard-coded
const Image = () => (
  <img
    src="https://images-na.ssl-images-amazon.com/images/I/517h-u1AQlL._SX218_BO1,204,203,200_QL40_FMwebp_.jpg"
    alt="book cover"
  />
)

""")
```

```python
# Syntax for variable constructor

Javascript(""" 

const Book = () => {
  const title = "I Love you to the Moon and Back"  
  const author = "Amelia Hepworth"
  return (<article className='book'>
      <Image />
      <h1>{title}</h1>      // soft coded; allows for a more flexible value control. 
      <h4>{author}</h4>     // soft coded
    </article>)
}

const Image = () => (
  <img
    src="https://images-na.ssl-images-amazon.com/images/I/517h-u1AQlL._SX218_BO1,204,203,200_QL40_FMwebp_.jpg"
    alt="book cover"
  />
)

""")
```

### 3-1. Instances form `Props`

(**Instances from a constructor**)

Let's re-write the code to allow to create multiple instances out of the constructor.

```python
# Compare the following: 
# a variable with hard-coded values.
Javascript("""

const Book = () => {
  const title = "I Love you to the Moon and Back"
  const author = "Amelia Hepworth"
  return (<article className='book'>
      <Image />
      <h1>{title}</h1>
      <h4>{author.toUpperCase()}</h4>
      <p>{ 5354 + 22}</p>
    </article>)
}

const Image = () => (
  <img
    src="https://images-na.ssl-images-amazon.com/images/I/517h-u1AQlL._SX218_BO1,204,203,200_QL40_FMwebp_.jpg"
    alt="book cover"
  />
)

""")
```

```python
# Use the power of JavaScript. 
# reorganise the code into 1) constructor, then 2) instances (vars), then call them in the Booklist.

Javascript("""

// constructor - Book
const Book = (props) => {
  console.log(props);
  return (
    <article className='book'>
      <img src={props.img} alt='' /> 
      <h1>{props.title}</h1>
      <h4>{props.author}</h4>
    </article>
  );
}

// setup vars
const Book1 = {
  img: "https://images-na.ssl-images-amazon.com/images/I/517h-u1AQlL._SX218_BO1,204,203,200_QL40_FMwebp_.jpg",
  title: 'I Love you to the Moon and Back',
  author: "Amelia Hepworth"
}

function Booklist() {
  return <section className='booklist'>
      <Book img={Book1.img} title={Book1.title} author={Book1.author}/>
      <Book img={Book2.img} title={Book2.title} author={Book2.author}/>
      ...
    </section>
  }

""")
```

### 3-2. `Prop` destructuring

```python
# Before destructuring
Javascript("""

const Book = (props) => {
  return (
    <article className='book'>
      <img src={props.img} alt='' /> 
      <h1>{props.title}</h1>
      <h4>{props.author}</h4>
    </article>
  );
}

""")
```

##### Destructuring syntax 1

```python
# Destructured props
Javascript("""

// constructor - Book
const Book = (props) => {
  const {img, title, author} = props;   // destructuring
  return (
    <article className='book'>
      <img src={img} alt='' /> 
      <h1>{title}</h1>
      <h4>{author}</h4>
    </article>
  );
}

""")
```

##### Destructuring syntax 2

```python
Javascript("""

const Book = ({ img, title, author}) => {   // destructuring
  return (
    <article className='book'>
      <img src={img} alt='' /> 
      <h1>{title}</h1>
      <h4>{author}</h4>
    </article>
  );
}

""")
```
